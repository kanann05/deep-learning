```python
# üß† STEP 2: Import required libraries
from ultralytics import YOLO
import cv2
from matplotlib import pyplot as plt
```

```python

# üöÄ STEP 3: Load a pretrained YOLOv8 model
# (YOLOv8s is small & fast ‚Äî ideal for beginners)
model = YOLO("yolov8s.pt")  # Downloads automatically
```

```python
# üñºÔ∏è STEP 4: Load an image for detection
# You can replace this with any image path or URL
image_path = "/content/1900-151662242_tiny.jpg"
```

```python

# STEP 5: Run object detection
results = model(image_path)

# STEP 6: Display detection results
# YOLO automatically saves results with bounding boxes
results[0].show()  # Opens window (on local) OR use below for Colab
```

```python
# ‚úÖ For Google Colab / Jupyter display
result_image = results[0].plot()
plt.figure(figsize=(10,10))
plt.imshow(cv2.cvtColor(result_image, cv2.COLOR_BGR2RGB))
plt.axis("off")
plt.show()
```

```python

# (Optional) STEP 7: Print detected objects
for box in results[0].boxes:
    cls = int(box.cls[0])
    conf = float(box.conf[0])
    name = model.names[cls]
    print(f"{name} - {conf:.2f}")
```
